```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(
    "GSE75688_GEO_processed_Breast_Cancer_raw_TPM_matrix.txt", delimiter="\t")
```

```python
data = df.T
```

```python
pooled_cells = data.iloc[3:17]
pooled_cells.columns = df['gene_name']
```

```python
sample_cells = data.iloc[17:]
sample_cells.columns = df['gene_name']
```

```python
all_cells = data.iloc[3:]
all_cells.columns = df['gene_name']
all_cells['gene_name'] = all_cells.index
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(sample_cells)
plt.figure(figsize=(10, 10))
sns.scatterplot(X_pca)
plt.show()
```
